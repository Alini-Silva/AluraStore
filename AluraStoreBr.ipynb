#1. Análise do faturamento


```python
faturamento_por_loja = {
    'Loja 1': loja['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

faturamento = {loja: float(valor) for loja, valor in faturamento_por_loja.items()}
print(faturamento)
```

```python
loja_menor_faturamento = min(faturamento, key=faturamento.get)
menor_valor = faturamento[loja_menor_faturamento]
print(f"A loja com menor faturamento é {loja_menor_faturamento} com R$ {menor_valor:.2f}")
```

```python
media_faturamento = pd.Series(faturamento_por_loja).mean()
print(f'{media_faturamento:.2f}')
```

```python
valores_desvio = []
for i in valores:
  desvio = abs(media_faturamento - i)
  valores_desvio.append(desvio)

valores_desvio
```

```python
def autolabel(loja):
  for i in loja:
    h = i.get_height()
    ax.annotate('{}'.format(h),
                xy = (i.get_x()+i.get_width()/2,h),
                xytext = (0,3),
                textcoords = 'offset points',
                ha = 'center')
```

##Objetivo
A presente análise tem como finalidade identificar a loja com o menor desempenho, a fim de subsidiar a decisão de encerramento de uma unidade. Foram considerados os seguintes critérios: faturamento total, categorias mais populares, média de avaliação dos clientes, produtos mais e menos vendidos e custo médio do frete.


---


##Resultados
Iniciando pela análise de faturamento, a Loja 4 obteve o menor valor, com 1.384.497,58. Em comparação, a Loja 1 obteve 1.534.509,12. O faturamento médio entre todas as lojas foi de 1.467.872,70, o que coloca a Loja 4 abaixo da média por 83.375,12.
No quesito avaliação, a Loja 4 mantém um bom desempenho, com média semelhante às demais lojas. Portanto, esse fator não é decisivo para a escolha da loja a ser encerrada.
Quanto às categorias de produtos, observa-se que "Móveis" é a mais vendida em todas as lojas. No entanto, a Loja 4 não apresenta destaque em nenhuma outra categoria relevante, ao contrário das demais. Por exemplo, nas Lojas 1 e 2, "Utilidades Domésticas" têm baixa procura, enquanto nas Lojas 3 e 4, "Instrumentos Musicais" são os menos vendidos. A ausência de uma categoria secundária forte na Loja 4 pode indicar falta de diversidade de apelo comercial.
Em relação ao frete médio, a Loja 4 apresenta o menor custo (31,28), seguida pela Loja 3 (33,07), Loja 2 (33,62) e Loja 1 (34,69). Apesar de ser uma vantagem competitiva, esse fator não foi suficiente para alavancar o faturamento da Loja 4.


---


##Conclusão
Portanto, após analisar todos os dados, conclui-se que a Loja 4 apresenta o desempenho mais fraco, tanto em termos de faturamento quanto de destaque em categorias de produto.
